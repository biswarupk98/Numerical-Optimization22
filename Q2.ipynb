Q2. Steepest Descent Gradient on f1(x) with initial x0=(2,0)

```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt

x0=tf.Variable([[2.0],[0.0]])
x=tf.Variable(x0)
xdata=x0
A=tf.Variable([[1.0,0.0],[0.0,1.0]])
def f(x,A):
    return tf.multiply(tf.matmul(tf.transpose(x),tf.matmul(A,x)),0.5)

rho=0.9
c=10**(-4)
itr=0
zdata=f(x0,A)

while(itr>=0):
    alpha=0.4
    x0=tf.Variable(x)
    with tf.GradientTape() as t:
        fx = tf.multiply(tf.matmul(tf.transpose(x0),tf.matmul(A,x0)),0.5)
    df = t.gradient(fx,x0)

    while(f(x0-tf.multiply(alpha,df),A)>=f(x0,A)-tf.multiply(c*alpha,tf.matmul(tf.transpose(df),df))):
        alpha=rho*alpha

    x=x0-tf.multiply(alpha,df)
    itr+=1
    xdata=tf.concat([xdata,x],1)
    zdata=tf.concat([zdata,f(x,A)],1)
    if(tf.norm(x-x0,ord='euclidean')<10**(-5)):
      break

print("minima at = ",x)
print("function value at minima = ",f(x,A))
print("number of iteration =",itr)

X,Y=np.meshgrid(np.linspace(-5.0,5.0,100),np.linspace(-5.0,5.0,100))
Z=(np.square(X)+np.square(Y))*0.5
cp=plt.contour(X,Y,Z)
plt.clabel(cp, inline=1, fontsize=10)

plt.plot(xdata[0,:],xdata[1,:],marker='.')
plt.show()
```

Q2. Newton's on f1(x) with initial x0=(2,0)

```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt

x0=tf.Variable([[2.0],[0.0]])
x=tf.Variable(x0)
xdata=x0
A=tf.Variable([[1.0,0.0],[0.0,1.0]])
def f(x,A):
    return tf.multiply(tf.matmul(tf.transpose(x),tf.matmul(A,x)),0.5)

rho=0.9
c=10**(-4)
itr=0
zdata=f(x0,A)

while(itr>=0):
    alpha=0.4
    x0=tf.Variable(x)
    
    with tf.GradientTape() as t2:
      with tf.GradientTape() as t1: 
        y =tf.multiply(tf.matmul(tf.transpose(x0),tf.matmul(A,x0)),0.5)
      df = t1.gradient(y, x0) 
    d2f = t2.jacobian(df, x0)

    d2f=tf.reshape(d2f,[2,2])
    while(f(x0-tf.multiply(alpha,df),A)>=f(x0,A)-tf.multiply(c*alpha,tf.matmul(tf.transpose(df),df))):
        alpha=rho*alpha

    x=x0-tf.multiply(alpha,tf.matmul(tf.linalg.inv(d2f),df))
    itr+=1
    xdata=tf.concat([xdata,x],1)
    zdata=tf.concat([zdata,f(x,A)],1)
    if(tf.norm(x-x0,ord='euclidean')<10**(-5)):
      break

print("minima at = ",x)
print("function value at minima = ",f(x,A))
print("number of iteration =",itr)

X,Y=np.meshgrid(np.linspace(-5.0,5.0,100),np.linspace(-5.0,5.0,100))
Z=(np.square(X)+np.square(Y))*0.5
cp=plt.contour(X,Y,Z)
plt.clabel(cp, inline=1, fontsize=10)

plt.plot(xdata[0,:],xdata[1,:],marker='.')
plt.show()
```
